```python
import cv2
```

```python
video = cv2.VideoCapture(0)
classfier = cv2.CascadeClassifier("cascade/cascade_caneca3.xml")
```

```python
while True:
    
    conect, frame = video.read()
    cinza_imagem = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    deteccoes = classfier.detectMultiScale(cinza_imagem, scaleFactor=1.3, minNeighbors=9, minSize=(60, 60))
    
    for (x, y, l, a) in deteccoes:
        cv2.rectangle(frame, (x, y), (x + l, y + a), (0, 0, 255), 2)
    cv2.imshow("video", frame)
    
    if cv2.waitKey(1) == ord("q"):
        break
        
video.release()
cv2.destroyAllWindows()
```
